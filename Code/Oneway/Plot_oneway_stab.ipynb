## Plot out wavenumber dependance of stability

### import package

```python
import sys;
import h5py;
import numpy as np;
import matplotlib.pyplot as plt;

sys.path.append("/home/b11209013/Package/")
import Plot_Style as ps;
```

### load data

```python
with h5py.File("/home/b11209013/2025_Research/MSI/File/oneway_stability.h5") as f:
    kn     = f["kn"][:];
    modal  = f["modal"][:];
```

### Plot out stability curve

```python
ps.apply_custom_plot_style();
clist = ps.colorlist();

fig, ax1 = plt.subplots(figsize=(16, 9));
l1       = ax1.scatter(kn, modal/2, c=clist[0], label="Modal");
ax1.set_xticks(
    np.linspace(0, 30, 7),
    [" ", "5", "10", "15", "20", "25", "30"]
)
ax1.set_yticks(
    np.linspace(0.0, 0.1, 6),
    [" ", "0.02", "0.04", "0.06", "0.08", "0.10"]
)
ax1.set_xlim(0, 30)
ax1.set_ylim(0, 0.1)
ax1.set_xlabel("Wavenumber")
ax1.set_ylabel("Modal")

plt.tight_layout()
plt.savefig("/home/b11209013/2025_Research/MSI/Fig/Oneway_Stability.png", dpi=200)
```
